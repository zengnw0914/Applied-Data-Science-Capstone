**Summary of Charts Plotted and Their Purpose**

Flight Number vs Payload Mass (catplot):
Plotted to visualize the relationship between the order of flights and the payload mass, with points colored by launch success (Class). This helps identify if higher flight numbers or certain payload ranges are associated with more successful launches.

Flight Number vs Launch Site (catplot):
Plotted to show how the sequence of flights is distributed across different launch sites, again colored by success. This allows for comparison of site performance over time.

Payload Mass vs Launch Site (catplot):
Used to examine how payload mass varies by launch site and how it relates to launch success. This helps determine if certain sites handle heavier or lighter payloads more successfully.

Success Rate by Orbit Type (barplot):
Plotted the average success rate for each orbit type. This chart helps identify which orbit types are associated with higher or lower launch success.

Flight Number vs Orbit (catplot):
Visualizes the relationship between flight sequence and orbit type, colored by success. This can reveal trends in the types of orbits targeted over time and their associated success rates.

Payload Mass vs Orbit Type (scatterplot):
Plotted to see how payload mass varies with orbit type and how this affects launch success. This helps in understanding if certain orbits are more challenging for heavier payloads.

Yearly Launch Success Rate Trend (lineplot):
Shows the trend of average launch success rate over the years. This chart is useful for identifying improvements or changes in SpaceX's launch reliability over time.

**Why these charts were used:**

These charts were selected to explore relationships between flight characteristics (number, payload, site, orbit) and launch outcomes, to identify patterns in success rates, and to track improvements over time. Visualizing these relationships helps in understanding key factors influencing launch success and operational trends.

```python
!pip install numpy pandas seaborn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
df = pd.read_csv(URL)
df.head(5)
#print(df)
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)  # hue:colors the points according to the Class column
plt.xlabel("Flight Number",fontsize=20)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

**TASK 1: Visualize the relationship between Flight Number and Launch**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Task 1: Flight Number vs Launch Site
plt.figure(figsize=(16, 6))
sns.catplot(data=df, x='FlightNumber', y='LaunchSite', hue='Class', kind='strip', height=6, aspect=2)
plt.xlabel('Flight Number', fontsize=16)
plt.ylabel('Launch Site', fontsize=16)
plt.title('Flight Number vs Launch Site by Class', fontsize=18)
plt.show()
```

**TASK 2: Visualize the relationship between Payload Mass and Launch Site**

```python
# Task 2: Payload Mass vs Launch Site
# plt.figure(figsize=(16, 6))
sns.catplot(data=df, x='PayloadMass', y='LaunchSite', hue='Class', kind='strip', height=6, aspect=2)
plt.xlabel('Payload Mass (kg)', fontsize=16)
plt.ylabel('Launch Site', fontsize=16)
plt.title('Payload Mass vs Launch Site by Class', fontsize=18)
plt.show()
```

**TASK 3: Visualize the relationship between success rate of each orbit type**

```python
# Task 3: Success Rate by Orbit Type
orbit_success = df.groupby('Orbit')['Class'].mean().reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(data=orbit_success, x='Orbit', y='Class', palette='viridis')
plt.ylabel('Success Rate', fontsize=16)
plt.xlabel('Orbit Type', fontsize=16)
plt.title('Success Rate by Orbit Type', fontsize=18)
plt.show()
```

**TASK 4: Visualize the relationship between FlightNumber and Orbit type**

```python
# Task 4: Flight Number vs Orbit
plt.figure(figsize=(16, 6))
sns.catplot(data=df, x='FlightNumber', y='Orbit', hue='Class', kind='strip', height=6, aspect=2)
plt.xlabel('Flight Number', fontsize=16)
plt.ylabel('Orbit', fontsize=16)
plt.title('Flight Number vs Orbit by Class', fontsize=18)
plt.show()
```

**TASK 5: Visualize the relationship between Payload Mass and Orbit type¶**

```python
plt.figure(figsize=(16, 6))
sns.scatterplot(data=df, x='PayloadMass', y='Orbit', hue='Class', palette='Set2', s=100) #oakette: color  s = size of scatter point
plt.xlabel('Payload Mass (kg)', fontsize=16)
plt.ylabel('Orbit Type', fontsize=16)
plt.title('Payload Mass vs Orbit Type by Class', fontsize=18)
plt.show()
```

**TASK 6: Visualize the launch success yearly trend**

```python
# Extract year from the 'Date' column
year = []
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Year'] = year

# Calculate average success rate per year
yearly_success = df.groupby('Year')['Class'].mean().reset_index()

# Plot the trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_success, x='Year', y='Class', marker='o')
plt.xlabel('Year', fontsize=16)
plt.ylabel('Success Rate', fontsize=16)
plt.title('Yearly Launch Success Rate Trend', fontsize=18)
plt.xticks(rotation=45)
plt.show()
```
